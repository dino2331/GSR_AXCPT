```python
sub = participants[1]

from scipy.signal import butter, filtfilt
from scipy.signal import sosfilt, sosfiltfilt
import pandas as pd
import numpy as np
# ────────────────────────────────
# 1 밴드패스 함수 정의 (루프 밖, 한 번만)
# ────────────────────────────────

def bandpass_filter(signal, fs, low=0.05, high=1.0, order=4):
    # 길이·분산 체크 → 오류 예방
    if len(signal) < order * 3 or np.std(signal) < 1e-6:
        return signal          # 필터 생략

    nyq = 0.5 * fs
    if low <= 0 or high >= nyq or low >= high:
        return signal          # 잘못된 컷오프 방지

    sos = butter(order, [low/nyq, high/nyq],
                 btype="band", output="sos")
    try:
        return sosfiltfilt(sos, signal)
    except Exception:          # 남은 예외는 1회 sosfilt 로 fallback
        return sosfilt(sos, signal)

# Step 1.1: GSR 전처리
gsr_raw = pd.read_csv(local_paths[sub]["gsr"], sep="\t", skiprows=1)
gsr = gsr_raw[[
    "Shimmer_AD82_TimestampSync_Unix_CAL",
    "Shimmer_AD82_GSR_Skin_Conductance_CAL"
]].copy()
gsr.columns = ["Timestamp", "Conductance"]
gsr = gsr[pd.to_numeric(gsr["Timestamp"], errors="coerce").notna()]
gsr["Timestamp"] = gsr["Timestamp"].astype(float)
gsr["Conductance"] = pd.to_numeric(gsr["Conductance"], errors="coerce")
gsr.dropna(inplace=True)
gsr["Timestamp_sec"] = gsr["Timestamp"] / 1000.0

# Step 1.2: AX-CPT 반응 전처리
axcpt_raw = pd.read_csv(local_paths[sub]["axcpt"])
valid_axcpt = axcpt_raw[axcpt_raw["key_resp_4.keys"].notna()].copy()
valid_axcpt = valid_axcpt[[
    "thisN", "Word", "CRESP", "key_resp_4.keys", "key_resp_4.corr",
    "key_resp_4.rt", "taskCue.started", "taskCue.stopped", "expStart"
]]
valid_axcpt.columns = [
    "TrialIndex", "Word", "CorrectKey", "PressedKey", "Correct",
    "RT", "TaskStart", "TaskEnd", "ExpStart"
]
valid_axcpt[["TrialIndex", "Correct", "TaskStart", "ExpStart"]] = valid_axcpt[[
    "TrialIndex", "Correct", "TaskStart", "ExpStart"
]].apply(pd.to_numeric, errors="coerce")
valid_axcpt["RT"] = valid_axcpt["RT"].apply(lambda x: x[0] if isinstance(x, list) else (
    eval(x)[0] if isinstance(x, str) and x.startswith("[") else np.nan))
if valid_axcpt["ExpStart"].isna().all():
    valid_axcpt["ExpStart"] = valid_axcpt["TaskStart"].min()
valid_axcpt["TrialStart_sec"] = valid_axcpt["TaskStart"] - valid_axcpt["ExpStart"]
valid_axcpt["IsCorrect"] = valid_axcpt["Correct"] == 1

# Step 1.3: 마커 동기화
marker_df = pd.read_excel(local_paths[sub]["marker"])
marker_time = marker_df[marker_df["marker"] == 1001]["RealTime"].iloc[0]
gsr["Time_sec_from_marker"] = gsr["Timestamp_sec"] - marker_time
valid_axcpt["TrialStart_absolute"] = valid_axcpt["TaskStart"] + marker_time

# Step 1.4: 조건 분류
valid_axcpt["Cue"] = valid_axcpt["Word"].shift(1)
valid_axcpt["Target"] = valid_axcpt["Word"]
def classify_axcpt_condition(row):
    cue = row["Cue"]
    target = row["Target"]
    if cue == "A" and target == "X":
        return "AX"
    elif cue == "A" and target != "X":
        return "AY"
    elif cue != "A" and target == "X":
        return "BX"
    elif cue != "A" and target != "X":
        return "BY"
    else:
        return "Unknown"
valid_axcpt["Condition"] = valid_axcpt.apply(classify_axcpt_condition, axis=1)

# neurokit2가 설치되었다고 가정하고 최종 merged_df 구성 코드 작성 (변수명 기존 유지)

import neurokit2 as nk
from scipy.signal import butter, filtfilt

# Step 1.5: GSR feature 추출 (neurokit2 기반 SCR_Count)
SCR_WINDOW_START = 1.0
SCR_WINDOW_END = 4.0

epoch_feature_list = []

for i, row in valid_axcpt.iterrows():
    trial_id = int(row["TrialIndex"])
    condition = row["Condition"]

    # Trial 기준 시간 범위
    start = row["TrialStart_absolute"] + SCR_WINDOW_START
    end = row["TrialStart_absolute"] + SCR_WINDOW_END

    gsr_segment = gsr[(gsr["Timestamp_sec"] >= start) & (gsr["Timestamp_sec"] <= end)].copy()
    conductance = gsr_segment["Conductance"].values

    scr_count = 0  # 기본값 설정

    if len(conductance) > 10:
        duration = gsr_segment["Timestamp_sec"].values[-1] - gsr_segment["Timestamp_sec"].values[0]
        original_sampling_rate = len(conductance) / duration if duration > 0 else 128
        if original_sampling_rate < 1:
            print(f"Trial {trial_id}: 샘플링 속도 너무 낮음 ({original_sampling_rate:.2f}Hz), 분석 제외")
        else:
          # 밴드 패스 필터링
            conductance = bandpass_filter(conductance,
                              fs=original_sampling_rate,
                              low=0.05, high=1.0, order=4)
            try:
                # 피부 전도도 데이터를 5Hz 주파수로 재추출 수행하기
                resampled_out = nk.signal_resample(conductance,
                                              sampling_rate=original_sampling_rate,
                                              desired_sampling_rate=5)

                if isinstance(resampled_out, tuple):       # v0.2.4+
                    resampled, info = resampled_out
                else:                                      # v0.2.3 이하
                    resampled = resampled_out

                sampling_rate = 5  # 샘플링 속도 강제 지정 5

                # 신호 필터링
                eda_cleaned = nk.eda_clean(resampled, sampling_rate=sampling_rate)

                # tonic/phasic 분리
                eda_df = nk.eda_phasic(eda_cleaned, sampling_rate=sampling_rate)
                # print(type(eda_df))
                # print(eda_df.head())

                if isinstance(eda_df, pd.DataFrame) and "EDA_Phasic" in eda_df.columns:
                    eda_phasic_value = eda_df["EDA_Phasic"].values
                    # print('eda형태 최종', eda_phasic_value.shape, 'sampling rate', sampling_rate)
                    # print('eda 내용 5개', eda_phasic[0:5])
                    # SCR 피크 탐지
                    scr_info = nk.eda_findpeaks(
                        eda_phasic_value,
                        sampling_rate=sampling_rate,
                        amplitude_min=0.05,
                    )
                    scr_count = scr_info["SCR_Peaks"].size # 키	내용 (dtype)	설명 "SCR_Peaks"	np.ndarray (정수 인덱스)
                    #	피크가 일어난 시점의 인덱스 목록입니다. 0/1 마스크나 개수(count)가 아니라, 보간-된 phasic 신호(약 1000 Hz 내부 해상도)의 샘플 번호가 들어 있습니다.
                    # 따라서 피크 개수를 찾을려면 이들을 더하는 것이 아닌 개수인 size를 찾으면 된다.
                else:
                    print(f"Trial {trial_id}: eda_phasic_df 구조 이상 → 분석 제외")
                    print("eda_phasic_df 내용:", eda_df)
            except Exception as e:
                print(f"Trial {trial_id}: NeuroKit 처리 실패 → {e}")
    else:
        print(f" Trial {trial_id}: conductance 길이 부족 (len={len(conductance)})")

    # Feature 저장
    if len(conductance) > 1:
        epoch_feature_list.append({
            "TrialIndex": trial_id,
            "Condition": condition,
            "MeanConductance": np.mean(conductance),
            "MaxConductance": np.max(conductance),
            "StdConductance": np.std(conductance),
            "DeltaConductance": conductance[-1] - conductance[0],
            "SCR_Count": scr_count
        })

# DataFrame으로 정리
gsr_features_df = pd.DataFrame(epoch_feature_list)

# 병합: 기존 변수명 유지
gsr_features_df_no_cond = gsr_features_df.drop(columns=["Condition"])
merged_df = pd.merge(valid_axcpt, gsr_features_df_no_cond, on="TrialIndex", how="inner")
print(merged_df.head())
```

```python
# 1. 필요 패키지
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 2. BSI 변화량 기반 그룹 설정
pair_bsi_df["BSI_Change"] = pair_bsi_df["Second_BSI"] - pair_bsi_df["First_BSI"]

increase_group = pair_bsi_df[pair_bsi_df["BSI_Change"] > 0]["subID"].tolist()
decrease_group = pair_bsi_df[pair_bsi_df["BSI_Change"] < 0]["subID"].tolist()

# 3. 추가 정보 병합
demo = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/GSR PJ final/GSR_DATA/demography.xlsx")
demo = demo[["subID", "BDI", "BAI", "LC 점수", "RC 점수"]]

summary_per_sub = merged_df.groupby("subID").agg({
    "RT": "mean",
    "IsCorrect": "mean",
    "SCR_Count": "mean"
}).reset_index()

full_summary = pd.merge(summary_per_sub, demo, on="subID", how="inner")

full_summary["BSI_Change_Group"] = full_summary["subID"].apply(lambda x:
                                       "Increase" if x in increase_group else
                                       ("Decrease" if x in decrease_group else np.nan))
full_summary = full_summary.dropna(subset=["BSI_Change_Group"])

# 4. 분석할 변수 목록
analysis_vars = ["RT", "IsCorrect", "SCR_Count", "BDI", "BAI", "LC 점수", "RC 점수"]

# 5. 그룹별 비교 + 통계
results = {}

for var in analysis_vars:
    group1 = full_summary[full_summary["BSI_Change_Group"] == "Increase"][var]
    group2 = full_summary[full_summary["BSI_Change_Group"] == "Decrease"][var]

    # 정규성 검정 (Shapiro-Wilk)
    p1 = stats.shapiro(group1)[1]
    p2 = stats.shapiro(group2)[1]

    # 정규성 만족 여부 확인
    if p1 > 0.05 and p2 > 0.05:
        # 독립 표본 t-test
        t_stat, p_val = stats.ttest_ind(group1, group2)
        test_used = "t-test"
    else:
        # Mann-Whitney U-test
        t_stat, p_val = stats.mannwhitneyu(group1, group2, alternative="two-sided")
        test_used = "Mann-Whitney"

    results[var] = {
        "group1_mean": group1.mean(),
        "group2_mean": group2.mean(),
        "statistic": t_stat,
        "p-value": p_val,
        "test": test_used
    }

# 6. 결과 출력
print("=== BSI 증가 vs 감소 그룹 비교 결과 ===")
for var, res in results.items():
    print(f"{var}: {res['test']} | t/U = {res['statistic']:.4f}, p = {res['p-value']:.4f}")

# 7. 시각화
for var in analysis_vars:
    plt.figure(figsize=(7,5))
    sns.violinplot(data=full_summary, x="BSI_Change_Group", y=var, inner="box", palette="pastel")
    plt.title(f"{var} by BSI Change Group")
    plt.ylabel(var)
    plt.xlabel("BSI Change Group")
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Trial별로 다음 trial 정보 추가
# -----------------------------
# TrialIndex 기준 정렬
merged_df = merged_df.sort_values(by=["subID", "TrialIndex"]).reset_index(drop=True)

# 다음 Trial의 정보 추가
merged_df["Next_RT"] = merged_df.groupby("subID")["RT"].shift(-1)
merged_df["Next_SCR"] = merged_df.groupby("subID")["SCR_Count"].shift(-1)

# -----------------------------
# 2. Conflict / Non-Conflict 조건 분리
# -----------------------------
# 현재 trial이 AY이면 다음 trial은 post_AY
merged_df["PostConflictType"] = np.where(merged_df["Condition"] == "AY", "post_AY", "post_nonAY")

# 단, 마지막 trial은 무조건 제외 (Next_RT가 NaN인 것 제거)
valid_df = merged_df.dropna(subset=["Next_RT", "Next_SCR"]).copy()

# -----------------------------
# 3. 그룹별 데이터 분리
# -----------------------------
post_ay_rt = valid_df[valid_df["PostConflictType"] == "post_AY"]["Next_RT"]
post_nonay_rt = valid_df[valid_df["PostConflictType"] == "post_nonAY"]["Next_RT"]

post_ay_scr = valid_df[valid_df["PostConflictType"] == "post_AY"]["Next_SCR"]
post_nonay_scr = valid_df[valid_df["PostConflictType"] == "post_nonAY"]["Next_SCR"]

# -----------------------------
# 4. 통계 검정
# -----------------------------
# RT 비교 (쌍체 t-test or Wilcoxon)
rt_t_stat, rt_p_value = stats.ttest_ind(post_ay_rt, post_nonay_rt, equal_var=False)  # 독립 샘플 t-test
# SCR 비교
scr_t_stat, scr_p_value = stats.ttest_ind(post_ay_scr, post_nonay_scr, equal_var=False)

print("=== Conflict Adaptation 분석 결과 (전체 데이터 기준) ===")
print(f"[RT] post_AY vs post_nonAY: t = {rt_t_stat:.4f}, p = {rt_p_value:.4f}")
print(f"[SCR_Count] post_AY vs post_nonAY: t = {scr_t_stat:.4f}, p = {scr_p_value:.4f}")

# -----------------------------
# 5. 시각화
# -----------------------------
# RT 시각화
plot_rt = pd.DataFrame({
    "Condition": ["post_AY"] * len(post_ay_rt) + ["post_nonAY"] * len(post_nonay_rt),
    "RT": pd.concat([post_ay_rt, post_nonay_rt])
})

plt.figure(figsize=(8, 5))
sns.violinplot(data=plot_rt, x="Condition", y="RT", inner="box", palette="pastel")
plt.title("RT after AY vs non-AY Trials")
plt.ylabel("Response Time (sec)")
plt.grid(True)
plt.tight_layout()
plt.show()

# SCR 시각화
plot_scr = pd.DataFrame({
    "Condition": ["post_AY"] * len(post_ay_scr) + ["post_nonAY"] * len(post_nonay_scr),
    "SCR_Count": pd.concat([post_ay_scr, post_nonay_scr])
})

plt.figure(figsize=(8, 5))
sns.violinplot(data=plot_scr, x="Condition", y="SCR_Count", inner="box", palette="Set2")
plt.title("SCR Count after AY vs non-AY Trials")
plt.ylabel("Number of SCR Peaks")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------
# 1. AY 이후와 non-AY 이후 trial 나누기
# --------------------------

merged_df = merged_df.sort_values(by=['subID', 'TrialIndex'])
merged_df['PrevCondition'] = merged_df['Condition'].shift(1)
merged_df['PrevSubID'] = merged_df['subID'].shift(1)

# 세션이 바뀌지 않은 경우만 고려
same_subject = merged_df['subID'] == merged_df['PrevSubID']

# AY 이후 trial: 이전 trial이 AY였던 경우
post_ay = merged_df[(merged_df['PrevCondition'] == 'AY') & same_subject]
# non-AY 이후 trial: 이전 trial이 AY가 아니었던 경우
post_nonay = merged_df[(merged_df['PrevCondition'] != 'AY') & same_subject]

# --------------------------
# 2. 정확도 비교
# --------------------------

# 정답률 평균 계산
acc_post_ay = post_ay['IsCorrect'].mean()
acc_post_nonay = post_nonay['IsCorrect'].mean()

# 두 그룹 비교 (비모수 검정: Mann-Whitney U-test)
stat, p = stats.mannwhitneyu(post_ay['IsCorrect'], post_nonay['IsCorrect'])

print("=== AY 이후 vs non-AY 이후 정확도 비교 ===")
print(f"AY 이후 정답률 평균: {acc_post_ay:.4f}")
print(f"non-AY 이후 정답률 평균: {acc_post_nonay:.4f}")
print(f"Mann-Whitney U-test 결과: U = {stat:.4f}, p = {p:.4f}")

# --------------------------
# 3. 시각화
# --------------------------

# 데이터 준비
data_acc = pd.DataFrame({
    'Condition': ['post_AY'] * len(post_ay) + ['post_nonAY'] * len(post_nonay),
    'Accuracy': pd.concat([post_ay['IsCorrect'], post_nonay['IsCorrect']], axis=0)
})

# 바이올린 플롯
plt.figure(figsize=(8,5))
sns.violinplot(data=data_acc, x='Condition', y='Accuracy', inner='box', palette='pastel')
plt.title('Accuracy after AY vs non-AY Trials')
plt.ylim(-0.1, 1.1)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────
# 1. 참가자별 BSI_RT, d_prime_context 계산
# ─────────────────────────────────────────────
participant_list = merged_df['subID'].unique()

strategy_list = []

for sub in participant_list:
    sub_df = merged_df[merged_df['subID'] == sub]

    # AY, BX 반응속도 평균
    ay_rt = sub_df[sub_df['Condition'] == 'AY']['RT'].mean()
    bx_rt = sub_df[sub_df['Condition'] == 'BX']['RT'].mean()
    bsi_rt = ay_rt - bx_rt  # BSI_RT 정의

    # d' 계산을 위해 hit rate, false alarm rate 계산
    # - AX 조건에서 맞으면 hit
    # - BX 조건에서 틀리면 false alarm
    n_ax = len(sub_df[sub_df['Condition'] == 'AX'])
    n_bx = len(sub_df[sub_df['Condition'] == 'BX'])
    hit = sub_df[(sub_df['Condition'] == 'AX') & (sub_df['IsCorrect'])].shape[0]
    fa = sub_df[(sub_df['Condition'] == 'BX') & (~sub_df['IsCorrect'])].shape[0]

    # rate 계산 (0이나 1일 경우 보정)
    hit_rate = (hit + 0.5) / (n_ax + 1) if n_ax > 0 else np.nan
    fa_rate = (fa + 0.5) / (n_bx + 1) if n_bx > 0 else np.nan

    if not np.isnan(hit_rate) and not np.isnan(fa_rate):
        d_prime = stats.norm.ppf(hit_rate) - stats.norm.ppf(fa_rate)
    else:
        d_prime = np.nan

    strategy_list.append({
        'subID': sub,
        'BSI_RT': bsi_rt,
        'd_prime_context': d_prime
    })

df_strategy_summary = pd.DataFrame(strategy_list)

print("참가자별 전략 요약 테이블:")
print(df_strategy_summary.head())

# ─────────────────────────────────────────────
# 2. demo와 병합 (LC, RC 점수 포함)
# ─────────────────────────────────────────────
# demo 파일에 subID, LC 점수, RC 점수 존재
merged_score_df = pd.merge(df_strategy_summary, demo[['subID','LC 점수','RC 점수']], on='subID')

print("\n전략 요약 + TOEIC 점수 병합 결과:")
print(merged_score_df.head())

# ─────────────────────────────────────────────
# 3. 가설 1: LC vs BSI_RT
# ─────────────────────────────────────────────
r_lc_bsi, p_lc_bsi = stats.pearsonr(merged_score_df['LC 점수'], merged_score_df['BSI_RT'])
rho_lc_bsi, sp_p_lc_bsi = stats.spearmanr(merged_score_df['LC 점수'], merged_score_df['BSI_RT'])

print(f"\n[LC vs BSI_RT] Pearson r = {r_lc_bsi:.3f}, p = {p_lc_bsi:.3f}")
print(f"[LC vs BSI_RT] Spearman rho = {rho_lc_bsi:.3f}, p = {sp_p_lc_bsi:.3f}")

plt.figure(figsize=(6,5))
sns.regplot(data=merged_score_df, x='LC 점수', y='BSI_RT', ci=95, scatter_kws={'s':50})
plt.title('LC Score vs BSI_RT (Proactive)')
plt.axhline(0, linestyle='--', color='gray')
plt.tight_layout()
plt.show()

# ─────────────────────────────────────────────
# 4. 가설 2: RC vs d_prime_context
# ─────────────────────────────────────────────
r_rc_dprime, p_rc_dprime = stats.pearsonr(merged_score_df['RC 점수'], merged_score_df['d_prime_context'])
rho_rc_dprime, sp_p_rc_dprime = stats.spearmanr(merged_score_df['RC 점수'], merged_score_df['d_prime_context'])

print(f"\n[RC vs d'] Pearson r = {r_rc_dprime:.3f}, p = {p_rc_dprime:.3f}")
print(f"[RC vs d'] Spearman r = {rho_rc_dprime:.3f}, p = {sp_p_rc_dprime:.3f}")

plt.figure(figsize=(6,5))
sns.regplot(data=merged_score_df, x='RC 점수', y='d_prime_context', ci=95, scatter_kws={'s':50}, color='orange')
plt.title("RC Score vs d' (Context Sensity)")
plt.axhline(0, linestyle='--', color='gray')
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# 참가자별 SCR 반응률 기반 통합 분석 코드
# -------------------------------

import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 참가자별 평균 SCR 반응률 계산
def compute_scr_rate(merged_df):
    scr_summary = (
        merged_df.groupby('subID')['SCR_Binary']
        .mean()
        .reset_index()
        .rename(columns={'SCR_Binary': 'SCR_Rate'})
    )
    return scr_summary

# 2. 참가자별 SCR 비율과 기존 merged_score_df 연결
def merge_scr_with_scores(scr_summary, merged_score_df):
    merged_scr_score = pd.merge(scr_summary, merged_score_df, on='subID', how='inner')
    return merged_scr_score

# 3. Spearman 상관 분석 함수
def spearman_correlation(x, y):
    rho, pval = stats.spearmanr(x, y)
    return rho, pval

# 4. 분석 및 출력 수행
def scr_rate_analysis(merged_df, merged_score_df):
    # SCR 비율 계산
    scr_summary = compute_scr_rate(merged_df)

    # SCR 비율과 참가자별 요약 점수 연결
    merged_scr_score = merge_scr_with_scores(scr_summary, merged_score_df)

    print("\n=== 참가자별 SCR 반응률 + 요약 테이블 ===")
    print(merged_scr_score.head())

    print("\n=== Spearman 상관 분석 결과 (SCR_Rate 기준) ===")
    for var in ['LC 점수', 'RC 점수', 'BSI_RT', 'd_prime_context']:
        rho, pval = spearman_correlation(merged_scr_score['SCR_Rate'], merged_scr_score[var])
        print(f"▶ SCR_Rate vs {var}: rho = {rho:.3f}, p = {pval:.4f}")

    # 시각화 추가 (선택)
    plt.figure(figsize=(8, 5))
    sns.regplot(data=merged_scr_score, x='SCR_Rate', y='LC 점수', lowess=True)
    plt.title('SCR_Rate vs LC 점수')
    plt.xlabel('SCR 반응률')
    plt.ylabel('LC 점수')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# -------------------------------
# 실제 사용 예시 (merged_df, merged_score_df 가 준비되어 있다고 가정)
# -------------------------------

scr_rate_analysis(merged_df, merged_score_df)

# 주석처리한 마지막 줄은 준비 완료되었을 때 실행하면 돼!
```

```python
# 분석 함수 (SCR_Rate + DeltaConductance 두 기준 모두)
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def gsr_language_integration_analysis(merged_df, merged_score_df):
    # 1. 참가자별 SCR 비율 계산
    scr_summary = merged_df.groupby('subID').agg(
        SCR_Rate=('SCR_Binary', 'mean'),
        Mean_DeltaConductance=('DeltaConductance', 'mean')
    ).reset_index()

    # 2. 기존 merged_score_df와 병합
    merged = pd.merge(merged_score_df, scr_summary, on='subID')

    # 3. 스피어만 상관분석 수행
    result = {}
    targets = ['SCR_Rate', 'Mean_DeltaConductance']
    predictors = ['LC 점수', 'RC 점수', 'BSI_RT', 'd_prime_context']

    for target in targets:
        for predictor in predictors:
            rho, p = stats.spearmanr(merged[predictor], merged[target])
            result[f'{predictor} vs {target}'] = {'rho': rho, 'p-value': p}

    print("=== Spearman 상관 분석 결과 (SCR_Rate, DeltaConductance 기준) ===")
    for key, value in result.items():
        print(f"{key}: rho = {value['rho']:.3f}, p = {value['p-value']:.4f}")

    # 4. 간단한 시각화 추가
    for target in targets:
        for predictor in predictors:
            plt.figure(figsize=(6,4))
            sns.regplot(x=merged[predictor], y=merged[target], scatter_kws={'s':50}, line_kws={'color':'red'})
            plt.title(f"{predictor} vs {target}")
            plt.axhline(0, linestyle='--', color='gray')
            plt.axvline(0, linestyle='--', color='gray')
            plt.tight_layout()
            plt.show()

# 사용 예시
gsr_language_integration_analysis(merged_df, merged_score_df)
```

```python
from scipy.stats import ttest_ind

# 개인별, 조건별 SCR t-test를 수행하는 함수
def ttest_scr_by_individual(df):
    results = []

    for sub_id, sub_df in df.groupby("subID"):
        # 조건별 분할
        ax_scr = sub_df[sub_df["Condition"] == "AX"]["SCR_Count"]
        ay_scr = sub_df[sub_df["Condition"] == "AY"]["SCR_Count"]
        bx_scr = sub_df[sub_df["Condition"] == "BX"]["SCR_Count"]
        by_scr = sub_df[sub_df["Condition"] == "BY"]["SCR_Count"]

        # AX vs AY
        if len(ax_scr) > 1 and len(ay_scr) > 1:  # 표본 수가 1개 이상일 때만 수행
            t_stat_ay, p_val_ay = ttest_ind(ax_scr, ay_scr, equal_var=False)
        else:
            t_stat_ay, p_val_ay = np.nan, np.nan

        # AX vs BX
        if len(ax_scr) > 1 and len(bx_scr) > 1:
            t_stat_bx, p_val_bx = ttest_ind(ax_scr, bx_scr, equal_var=False)
        else:
            t_stat_bx, p_val_bx = np.nan, np.nan

        # AX vs BY
        if len(ax_scr) > 1 and len(by_scr) > 1:
            t_stat_by, p_val_by = ttest_ind(ax_scr, by_scr, equal_var=False)
        else:
            t_stat_by, p_val_by = np.nan, np.nan

        # 결과 저장
        results.append({
            "subID": sub_id,
            "AX_vs_AY_t": t_stat_ay, "AX_vs_AY_p": p_val_ay,
            "AX_vs_BX_t": t_stat_bx, "AX_vs_BX_p": p_val_bx,
            "AX_vs_BY_t": t_stat_by, "AX_vs_BY_p": p_val_by
        })

    ttest_df = pd.DataFrame(results)
    return ttest_df

# 함수 실행
scr_ttest_results = ttest_scr_by_individual(merged_df)

# 결과 미리 보기
scr_ttest_results.head()
```

```python
from scipy.stats import chi2_contingency

# 개인별 카이제곱 검정 함수
def chi_square_scr_binary(df):
    results = []

    for sub_id, sub_df in df.groupby("subID"):
        # 조건별 분할
        ax_bin = sub_df[sub_df["Condition"] == "AX"]["SCR_Binary"]
        ay_bin = sub_df[sub_df["Condition"] == "AY"]["SCR_Binary"]
        bx_bin = sub_df[sub_df["Condition"] == "BX"]["SCR_Binary"]
        by_bin = sub_df[sub_df["Condition"] == "BY"]["SCR_Binary"]

        def perform_chi2(x, y):
            # 관측 빈도 테이블 생성
            table = pd.DataFrame({
                "Condition1": [sum(x == 0), sum(x == 1)],
                "Condition2": [sum(y == 0), sum(y == 1)]
            })
            if table.values.min() >= 0 and table.values.sum() > 0:
                chi2, p, dof, expected = chi2_contingency(table)
            else:
                chi2, p = np.nan, np.nan
            return chi2, p

        # 각 비교 수행
        chi2_ay, p_ay = perform_chi2(ax_bin, ay_bin)
        chi2_bx, p_bx = perform_chi2(ax_bin, bx_bin)
        chi2_by, p_by = perform_chi2(ax_bin, by_bin)
        chi_bxy, p_bxy = perform_chi2(bx_bin, by_bin)

        results.append({
            "subID": sub_id,
            "AX_vs_AY_chi2": chi2_ay, "AX_vs_AY_p": p_ay,
            "AX_vs_BX_chi2": chi2_bx, "AX_vs_BX_p": p_bx,
            "AX_vs_BY_chi2": chi2_by, "AX_vs_BY_p": p_by,
            "BX_vs_BY_chi2": chi_bxy, "BX_vs_BY_p": p_bxy
        })

    chi2_df = pd.DataFrame(results)
    return chi2_df

# 함수 실행
scr_binary_chi2_results = chi_square_scr_binary(merged_df)

# 결과 미리보기
scr_binary_chi2_results
```

```python
# 참가자별 t-test와 카이제곱 검정 수행
def analyze_scr_effect_per_participant(df):
    results = []

    for sub_id, sub_df in df.groupby("subID"):
        # --- SCR_Binary vs RT (t-test) ---
        scr_0_rt = sub_df[sub_df["SCR_Binary"] == 0]["RT"]
        scr_1_rt = sub_df[sub_df["SCR_Binary"] == 1]["RT"]
        if len(scr_0_rt) > 1 and len(scr_1_rt) > 1:
            t_stat_rt, p_value_rt = ttest_ind(scr_0_rt, scr_1_rt, equal_var=False)
        else:
            t_stat_rt, p_value_rt = np.nan, np.nan

        # --- SCR_Binary vs Correct (카이제곱 검정) ---
        contingency_table = pd.crosstab(sub_df["SCR_Binary"], sub_df["Correct"])
        if contingency_table.shape == (2, 2):
            chi2_stat, p_value_chi2, dof, expected = chi2_contingency(contingency_table)
        else:
            chi2_stat, p_value_chi2 = np.nan, np.nan

        # 결과 저장
        results.append({
            "subID": sub_id,
            "SCR_RT_t_stat": t_stat_rt,
            "SCR_RT_p_value": p_value_rt,
            "SCR_Correct_chi2_stat": chi2_stat,
            "SCR_Correct_p_value": p_value_chi2
        })

    results_df = pd.DataFrame(results)
    return results_df

# 함수 실행
participant_analysis_results = analyze_scr_effect_per_participant(merged_df)

# 결과 미리 보기
participant_analysis_results
```

```python
# [가정] merged_df와 full_demo_df는 다음 컬럼을 가진 DataFrame으로 존재한다고 가정합니다.
# merged_df: subID, SCR_Binary
# full_demo_df: subID, sex (\uac00\ud559, \uc5ec\ud559 \ud574\uc11d \ud544\uc694)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, mannwhitneyu, shapiro

# 1. 참가자별 SCR_Rate 계산 (SCR_Binary 평균)
scr_summary = merged_df.groupby('subID')['SCR_Binary'].mean().reset_index()
scr_summary.columns = ['subID', 'SCR_Rate']

# 2. 성별 정보 가져오기 (full_demo_df)
sex_info = demo_df[['subID', 'sex']]
sex_info['sex'] = sex_info['sex'].replace({'남': 'M', '여': 'F'})  # 한글 \u2192 영어 통일

# 3. SCR_Rate와 성별 정보 병합
scr_summary = pd.merge(scr_summary, sex_info, on='subID')

# 4. 남성과 여성 그룹 나누기
male = scr_summary[scr_summary['sex'] == 'M']['SCR_Rate']
female = scr_summary[scr_summary['sex'] == 'F']['SCR_Rate']

# 5. 샘플 수 확인
n_male = len(male)
n_female = len(female)

# 6. 정규성 검사 (Shapiro-Wilk Test)
p_male_shapiro = shapiro(male).pvalue if n_male >= 3 else 1.0
p_female_shapiro = shapiro(female).pvalue if n_female >= 3 else 1.0

# 7. 통계기법 선택 및 검정
if p_male_shapiro > 0.05 and p_female_shapiro > 0.05:
    # 정규성 만족
    if abs(n_male - n_female) <= 5:
        # 샘플 수 비슷 \u2192 독립표본 t-test (equal_var=True)
        t_stat, p_val = ttest_ind(male, female, equal_var=True)
        test_used = "Independent t-test (equal_var=True)"
    else:
        # 샘플 수 차이 크다 \u2192 Welch's t-test (equal_var=False)
        t_stat, p_val = ttest_ind(male, female, equal_var=False)
        test_used = "Welch's t-test (equal_var=False)"
else:
    # 정규성 만족 못함 \u2192 Mann-Whitney U test
    t_stat, p_val = mannwhitneyu(male, female, alternative='two-sided')
    test_used = "Mann-Whitney U test"

# 8. 결과 출력
print("\n== 분석 결과 ==")
print(f"사용된 검정 방법: {test_used}")
print(f"남자 수: {n_male}, 여자 수: {n_female}")
print(f"남자 평균 SCR_Rate: {male.mean():.4f}")
print(f"여자 평균 SCR_Rate: {female.mean():.4f}")
print(f"검정 통계량: {t_stat:.4f}")
print(f"p-value: {p_val:.4f}")

# 9. 시각화 (Boxplot)
plt.figure(figsize=(6,5))
sns.boxplot(x='sex', y='SCR_Rate', data=scr_summary)
plt.title(f'Comparison of SCR response rates by gender ({test_used})')
plt.xlabel('sex')
plt.ylabel('SCR_Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# merged_df: subID, TrialStart_sec, RT, Correct
# full_demo_df: subID, BDI, FLRAS, FLCAS, BAI 등 불안 관련 점수

import pandas as pd
import numpy as np
from scipy.stats import linregress, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 각 참가자별 RT 변화량(slope) 계산
rt_slopes = []
for sub_id, group in merged_df.groupby("subID"):
    group = group.dropna(subset=["TrialStart_sec", "RT"])
    if len(group) >= 10:  # 최소 trial 수
        slope, intercept, r_value, p_value, std_err = linregress(group["TrialStart_sec"], group["RT"])
        rt_slopes.append({"subID": sub_id, "RT_slope": slope})

rt_slope_df = pd.DataFrame(rt_slopes)

# 2. (선택) 정확도 변화량(slope) 계산
acc_slopes = []
for sub_id, group in merged_df.groupby("subID"):
    group = group.dropna(subset=["TrialStart_sec", "Correct"])
    if len(group) >= 10:
        slope, intercept, r_value, p_value, std_err = linregress(group["TrialStart_sec"], group["Correct"])
        acc_slopes.append({"subID": sub_id, "Accuracy_slope": slope})

acc_slope_df = pd.DataFrame(acc_slopes)

# 3. 개인별 slope 병합
slope_summary = pd.merge(rt_slope_df, acc_slope_df, on="subID")

# 4. 불안 점수 가져오기 (full_demo_df)
anxiety_scores = demo_df[["subID", "BDI", "FLRAS", "FLCAS", "BAI"]]

# 5. slope 데이터와 불안 점수 병합
final_df = pd.merge(slope_summary, anxiety_scores, on="subID")

# 6. RT_slope vs 불안 점수 상관분석 (Spearman)
correlation_results = {}
for anxiety_var in ["BDI", "FLRAS", "FLCAS", "BAI"]:
    rho, pval = spearmanr(final_df["RT_slope"], final_df[anxiety_var])
    correlation_results[anxiety_var] = {"Spearman_rho": rho, "p_value": pval}

# 7. 결과 출력
print("\n== 개인별 RT 변화량과 불안 점수 간 상관 분석 결과 ==")
for var, result in correlation_results.items():
    print(f"{var}: rho = {result['Spearman_rho']:.4f}, p = {result['p_value']:.4f}")

# 8. 시각화 (예: BDI 점수 vs RT_slope)
plt.figure(figsize=(6,5))
sns.scatterplot(x=final_df["FLCAS"], y=final_df["RT_slope"])
sns.regplot(x=final_df["FLCAS"], y=final_df["RT_slope"], scatter=False, color='red')
plt.xlabel("FLCAS")
plt.ylabel("RT change (slope)")
plt.title("FLCAS score and RT ")
plt.grid(True)
plt.tight_layout()
plt.show()
```
